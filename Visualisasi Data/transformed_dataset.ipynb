```python
import pandas as pd

df_detail = pd.read_csv("/content/Details.csv")
df_orders = pd.read_csv("/content/Orders.csv")
```

```python
df_detail.head(10)
```

```python
df_orders.head(10)
```

```python
# Join two df on 'Order Id'
merged_df = pd.merge(df_detail, df_orders, on='Order ID', how='inner')

merged_df.head(10)
```

```python
#rename Amount column to Price
merged_df = merged_df.rename(columns={'Amount': 'Price'})
```

```python
# Count null values in each column
null_counts = merged_df.isnull().sum()
print("Null values per column:\n", null_counts)

# Remove rows with any null values
merged_df = merged_df.dropna()
```

```python
# Count duplicate rows based on all columns
duplicate_count = merged_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Remove duplicate rows, keeping the first occurrence
merged_dfs = merged_df.drop_duplicates()
```

```python
# 1. Calculate total amount for each order
merged_df['Total Price'] = merged_df['Quantity'] * merged_df['Price']
merged_df
```

```python
# Convert 'Order Date' column to datetime format
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], format='%d-%m-%Y')

merged_df
```
